```python
# set up python environment
import pandas as pd
import os
import numpy as np
import warnings
from pandas import read_csv
warnings.filterwarnings('ignore')

os.chdir('/Users/xiaodanxu/Library/CloudStorage/GoogleDrive-arielinseu@gmail.com/My Drive/GEMS/BILD-AQ')
state_list = ['AK', 'HI', 'OR', 'WA']
region_code = 'CD9'
```

```python
output_dist_matrix = None
for st in state_list:
    print(st)
    dist_mat = \
    pd.read_csv('data/Network/' + st + '/distance_matrix_by_tracts.csv')
    dist_mat = dist_mat.rename(columns = {'Unnamed: 0': 'origin'})
    dist_matrix_long = pd.melt(dist_mat, id_vars = 'origin', 
                               var_name = 'destination', value_name = 'distance')
    dist_matrix_long.loc[:, ['origin', 'destination']] = dist_matrix_long.loc[:, ['origin', 'destination']].astype(int)
    output_dist_matrix = pd.concat([output_dist_matrix, dist_matrix_long])

output_dist_matrix.to_csv('data/Network/combined/distance_matrix_by_tracts_' + region_code + '.csv')
output_dist_matrix.head(5)
```

```python
output_travel_skim = None
for st in state_list:
    print(st)
    travel_skim = \
    pd.read_csv('data/Network/' + st + '/travel_time_skim.csv')
    output_travel_skim = pd.concat([output_travel_skim, travel_skim])

output_travel_skim.to_csv('data/Network/combined/travel_time_skim_' + region_code + '.csv')
output_travel_skim.head(5)
```
