```python
# process shapefile
filepath = 'shapefile/tl_2017_06_tract/tl_2017_06_tract_no_island.geojson'
ca_census_tracts = gpd.read_file(filepath)
ca_census_tracts = ca_census_tracts.to_crs("epsg:4087")
centroids = np.column_stack((ca_census_tracts.centroid.x, 
                             ca_census_tracts.centroid.y))
# ca_census_tracts.head(5)
# geo_array = ca_census_tracts.geometry.explode()[1:].values
# ca_census_tracts.loc[:, 'geometry'] = gpd.GeoSeries([MultiPolygon(geo_array)])

queen = weights.Queen.from_dataframe(ca_census_tracts)
graph = queen.to_networkx()
```

```python
ax = ca_census_tracts.plot(figsize = (12, 8), linewidth=.8, alpha = 0.7)
cx.add_basemap(ax, crs = 'EPSG:4087', source = cx.providers.Stamen.TonerLite)
plt.savefig('shapefile/ca_tracts.png',dpi = 200)
plt.show()
```

```python
# define scenario/ analysis boundary
state = 'CA'
microtype_lookup = read_csv('data/ccst_geoid_key_tranps_geo_with_imputation.csv', sep = ',')
microtype_by_state = microtype_lookup.loc[microtype_lookup['st_code'] == state]
select_tracts = microtype_by_state.GEOID.unique()
print(len(select_tracts))
```

```python
route_url = 'data/route/output.csv'
chunk_size = 50000
selected_routes = None
i = 0
for chunk in read_csv(route_url, header = None, sep = ', ', 
                      chunksize = chunk_size, engine='python'):
    print('processing chunk ' + str(i))
    chunk.columns = ['origin', 'destination', 'geometry']
    chunk_selected = chunk.loc[chunk['origin'].isin(select_tracts)]
    chunk_selected = chunk_selected.loc[chunk_selected['destination'].isin(select_tracts)]
    selected_routes = pd.concat([selected_routes, chunk_selected])
    i += 1
# routes = pd.read_csv(route_url, header = None, sep = ', ', nrows = 500, engine='python')

selected_routes.head(5)
```

```python
selected_routes['geometry'] = selected_routes['geometry'].apply(shapely.wkt.loads)
routes_gdf = gpd.GeoDataFrame(selected_routes, crs='epsg:4326')
routes_gdf.head(1)
```

```python
ax = routes_gdf.plot(figsize = (12, 8), linewidth=.5, alpha = 0.5)
cx.add_basemap(ax, crs = 'EPSG:4326', source = cx.providers.Stamen.TonerLite)
plt.savefig(state + '_route.png',dpi = 200)
plt.show()
```

```python
print(len(routes_gdf))
routes_gdf.to_file('route/' + state + '_routes.geojson', driver='GeoJSON')
```

```python
# process second sources of routes
all_files = os.listdir("data/route")    
gz_files = list(filter(lambda f: f.endswith('.gz'), all_files))
print(gz_files)

select_tracts = select_tracts.astype(int)
chunk_size = 50000
for file in gz_files:
    filename = file.split('.gz')[0]
#     aws_route = None
    i = 0
    for chunk in read_csv("data/route/" + file, sep = ',', chunksize = chunk_size):
        outfile_name = filename + '_' + str(i) + '.geojson'
        if i % 10 == 0:
            print('processing chunk ' + str(i))
        chunk.columns = ['source', 'destination', 'distance', 'ascend', 'descend', 'geometry']
        chunk.loc[:, ['source', 'destination']] = chunk.loc[:, ['source', 'destination']].astype(int)
        chunk_selected = chunk.loc[chunk['source'].isin(select_tracts)]
        chunk_selected = chunk_selected.loc[chunk_selected['destination'].isin(select_tracts)]
#         aws_route = pd.concat([aws_route, chunk_selected])
        if len(chunk_selected) == 0:
            i += 1
            continue
        chunk_selected['geometry'] = chunk_selected['geometry'].apply(shapely.wkt.loads)
        routes2_gdf = gpd.GeoDataFrame(chunk_selected, crs='epsg:4326')
        routes2_gdf.to_file('data/route/processed/' + outfile_name, driver='GeoJSON')
#         routes2_gdf.head(5)
#         ax = routes2_gdf.plot(figsize = (12, 8), linewidth=.5, alpha = 0.5)
#         cx.add_basemap(ax, crs = 'EPSG:4326', source = cx.providers.Stamen.TonerLite)
#         plt.savefig('results/aws_sample_route_2.png',dpi = 200)
#         plt.show()       
        i += 1
#         break
```
