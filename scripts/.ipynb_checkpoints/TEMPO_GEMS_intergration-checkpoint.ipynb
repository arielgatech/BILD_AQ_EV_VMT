# Integrating TEMPO and GEMS output 
## Part 1 - set up scenario parameters and load inputs

```python
# load packages
import os
from pandas import read_csv
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import contextily as cx
import shapely.wkt
import matplotlib
import numpy as np

work_dir = '/Volumes/GoogleDrive/My Drive/GEMS/BILD-AQ/data'
os.chdir(work_dir)

plt.style.use('ggplot')
sns.set(font_scale=1.2)  # larger font
```

```python
# prepare input and parameters
analysis_year = 2018
scenario_name = '1.0' # this variable is currently hard coded, 
# will be ideal to load it from a config file
state_name = 'CA'
output_name = state_name + '_' + scenario_name + '_' + str(analysis_year)

link_to_tempo_output = 'Input/all_stock_tract_ca_0624.csv'
link_to_gems_output = 'Output/BILDAQ_VMT_by_tract.csv'
```

```python
# load GEMS output and compute VMT by home tract
gems_VMT = read_csv(link_to_gems_output, sep = ',', low_memory = 'False')
print(gems_VMT.columns)
gems_VMT.head(5)
```

```python
gems_VMT = gems_VMT.dropna(subset = 'home_GEOID')
# gems_VMT.loc[:, 'home_GEOID'] = gems_VMT.loc[:, 'home_GEOID'].astype(int)
gems_VMT.loc[:, 'home_GEOID'] = gems_VMT.loc[:, 'home_GEOID'].astype(str).str.zfill(11)
gems_VMT.loc[:, 'thru_GEOID'] = gems_VMT.loc[:, 'thru_GEOID'].astype(str).str.zfill(11)
home_VMT = gems_VMT.groupby('home_GEOID')[['VMT']].sum()
home_VMT = home_VMT.reset_index()
home_VMT.head(5)
```

```python
# load TEMPO output
tempo_result = read_csv(link_to_tempo_output, sep = ',')
tempo_result_summary = pd.pivot_table(tempo_result, values = 'Vehicles', 
                                      index = ['Year', 'Scenario'], columns = 'Tech2',
                                      aggfunc = np.sum)
print(tempo_result.columns)
tempo_result_summary = tempo_result_summary.reset_index()
tempo_result_summary.to_csv('Output/TEMPO_veh_count.csv', index = False)
```

```python
tempo_result_filtered = tempo_result.loc[tempo_result['Year'] == analysis_year]
tempo_result_filtered = tempo_result_filtered.loc[tempo_result_filtered['Scenario'] == scenario_name]
tempo_result_filtered.loc[:, 'Region'] = tempo_result_filtered.loc[:, 'Region'].astype(str).str.zfill(5)
tempo_result_filtered.loc[:, 'tractid'] = tempo_result_filtered.loc[:, 'tractid'].astype(str).str.zfill(6)
tempo_result_filtered.loc[:, 'home_GEOID'] = tempo_result_filtered.loc[:, 'Region'].astype(str) + \
tempo_result_filtered.loc[:, 'tractid']
home_veh = tempo_result_filtered.groupby('home_GEOID')[['Vehicles']].sum()
home_veh = home_veh.reset_index()
home_veh.head(5)
```

```python
# validation of GEMS results -- if high home VMT correlated with home veh ownership
home_VMT_vs_veh = pd.merge(home_VMT, home_veh, on = 'home_GEOID', how = 'inner')
ax = sns.scatterplot(x="Vehicles", y="VMT", data=home_VMT_vs_veh, alpha = 0.3)
ax.set(yscale="log")
plt.xlabel('vehicle count from TEMPO')
plt.ylabel('VMT from GEMS')
plt.savefig('Plot/gems_vmt_vs_tempo_veh.png', dpi = 200)
```

## Part 2 - allocate GEMS VMT by TEMPO vehicle composition

```python
# generate fleet composition by home tract
fleet_by_home_tract = pd.pivot_table(tempo_result_filtered, 
                                     values='Vehicles', index=['home_GEOID'],
                                     columns=['Tech2'])
list_of_veh_tech = fleet_by_home_tract.columns
fleet_by_home_tract.loc[:, 'Total'] = fleet_by_home_tract.loc[:, list_of_veh_tech].sum(axis = 1)
# fill tract with 0 vehicles, assume 1 hypothetical ICEV to ensure VMT can be assigned
fleet_by_home_tract.loc[fleet_by_home_tract['Total'] == 0, 'ICEV'] = 1
fleet_by_home_tract.loc[fleet_by_home_tract['Total'] == 0, 'Total'] = 1

for tech in list_of_veh_tech:
    #print(tech)
    fleet_by_home_tract.loc[:, tech] = fleet_by_home_tract.loc[:, tech] / fleet_by_home_tract.loc[:, 'Total']
fleet_by_home_tract = fleet_by_home_tract.reset_index()
fleet_by_home_tract.head(5)
```

```python
# combine fleet composition with GEMS VMT
gems_VMT_with_fleet_composition = pd.merge(gems_VMT, fleet_by_home_tract, 
                                           on = 'home_GEOID', how = 'inner')
for tech in list_of_veh_tech:
    #print(tech)
    gems_VMT_with_fleet_composition.loc[:, tech] = gems_VMT_with_fleet_composition.loc[:, tech] * \
    gems_VMT_with_fleet_composition.loc[:, 'VMT']
gems_VMT_with_fleet_composition.head(5)    
```

```python
list_attr = list_of_veh_tech.tolist()
list_attr.append('VMT')
print(list_attr)
VMT_by_thru_tract_and_veh = gems_VMT_with_fleet_composition.groupby(['thru_GEOID', 'thru_geotype', 'thru_microtype'])[list_attr].sum()

VMT_by_thru_tract_and_veh = VMT_by_thru_tract_and_veh.reset_index()
VMT_by_thru_tract_and_veh.to_csv('Output/VMT_' + output_name+ '.csv', index = False)
VMT_by_thru_tract_and_veh.head(5)
```

## Part 3 -- visualize EV ownership and EV VMT

```python
# generate map visualization
CA_tract_file = 'Network/CA_census_tracts_2010_clean.geojson'
CA_tracts_geojson =  gpd.read_file(CA_tract_file) # load shapefile of CA tracts
# CA_tracts_geojson.loc[:, 'home_GEOID'] = CA_tracts_geojson.loc[:, 'GEOID'].astype(int)
CA_tracts_geojson.loc[:, 'home_GEOID'] = CA_tracts_geojson.loc[:, 'GEOID'].astype(str).str.zfill(0)
CA_tracts_geojson.head(5)
CA_tracts_geojson.plot()
```

```python
# plot EV ownership by EV fraction
CA_tracts_with_EV = CA_tracts_geojson.merge(fleet_by_home_tract, 
                                            on = 'home_GEOID',
                                            how = 'left')
ax = CA_tracts_with_EV.plot(figsize = (10,6), column = 'BEV', cmap='coolwarm',
                            alpha = 0.6, linewidth=0.01, legend=True)
cx.add_basemap(ax, crs = 'EPSG:4326', source = cx.providers.Stamen.TonerLite)
plt.title('BEV fraction by home tract')
plt.savefig('Plot/TEMPO_BEV_frac_by_tract_alt.png', dpi = 200)
```

```python
# absolute BEV count
CA_tracts_with_EV.loc[:, 'BEV_count'] = CA_tracts_with_EV.loc[:, 'Total'] * CA_tracts_with_EV.loc[:, 'BEV']
ax = CA_tracts_with_EV.plot(figsize = (10,6), column = 'BEV_count', cmap='coolwarm',
                            alpha = 0.6, vmin=0, vmax=100, linewidth=0.0, legend=True)
cx.add_basemap(ax, crs = 'EPSG:4326', source = cx.providers.Stamen.TonerLite)
plt.title('BEV count by home tract')
plt.savefig('Plot/TEMPO_BEV_count_by_tract_alt.png', dpi = 200)
```

```python
# plot EV vmt per acre


sm_to_acre = 0.000247105
CA_tracts_with_VMT = CA_tracts_geojson.merge(VMT_by_thru_tract_and_veh, 
                                            left_on = 'home_GEOID',
                                             right_on = 'thru_GEOID',
                                            how = 'left')
CA_tracts_with_VMT.loc[:, 'acre'] = CA_tracts_with_VMT['geometry'].to_crs({'proj':'cea'}).area * sm_to_acre
CA_tracts_with_VMT.loc[:, 'BEV_per_acre'] = CA_tracts_with_VMT.loc[:, 'BEV'] / CA_tracts_with_VMT.loc[:, 'acre']
# CA_tracts_with_VMT.loc[:, 'BEV_per_acre'].hist(bins =30)
ax = CA_tracts_with_VMT.plot(figsize = (10,6), column = 'BEV_per_acre', cmap='coolwarm', 
                             alpha = 0.6, linewidth=0.01, vmax = 1,
#                              norm=matplotlib.colors.LogNorm(vmin=1, vmax=CA_tracts_with_VMT.BEV_per_acre.max()),
                             legend=True)
cx.add_basemap(ax, crs = 'EPSG:4326', source = cx.providers.Stamen.TonerLite)
plt.title('BEV VMT per acre by through tract')
plt.savefig('Plot/GEMS_BEV_vmt_by_acre_by_thru_tract_alt.png', dpi = 200)
```

```python
# BEV VMT 
ax = CA_tracts_with_VMT.plot(figsize = (10,6), column = 'BEV', cmap='coolwarm', 
                             alpha = 0.6, linewidth=0.01, 
                             norm=matplotlib.colors.LogNorm(vmin=10, vmax=CA_tracts_with_VMT.BEV.max()),
                             legend=True)
cx.add_basemap(ax, crs = 'EPSG:4326', source = cx.providers.Stamen.TonerLite)
plt.title('BEV VMT by through tract')
plt.savefig('Plot/GEMS_BEV_vmt_by_thru_tract_alt.png', dpi = 200)
```

```python
# BEV VMT fraction
CA_tracts_with_VMT.loc[:, 'BEV_fraction'] = CA_tracts_with_VMT.loc[:, 'BEV'] / CA_tracts_with_VMT.loc[:, 'VMT']
ax = CA_tracts_with_VMT.plot(figsize = (10,6), column = 'BEV_fraction', cmap='coolwarm', 
                             alpha = 0.6, linewidth=0.01, 
                             vmin = 0,
                             legend=True)
cx.add_basemap(ax, crs = 'EPSG:4326', source = cx.providers.Stamen.TonerLite)
plt.title('BEV VMT Fraction by through tract')
plt.savefig('Plot/GEMS_BEV_vmt_frac_by_thru_tract_alt.png', dpi = 200)
```
